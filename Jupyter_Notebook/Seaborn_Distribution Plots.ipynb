# Distribution Plots
## Understand the Plots types

* Distribution plots display a single continuous feature and help visualize properties such as deviation and average values
* There are 3 main distribution plot types:
    * Rug plot
    * Histogram
    * KDE Plot (Kernel Density Estimation)

***

## Rug plot
* Let's explore the distribution of employee salaries
* One way to do this through **a rug plot**
* **A rug plot** is the simplest distribution plot and merely adds **a dash** or **tick line** for every single value
* The y-axis does not really have a meaning
* Rug plot of salaries
    * Adds a tick for every salary value
    * Optionally adjust height of ticks
    * Y-axis is not interpretable
    * Mass of ticks represent for the main feature of distribution
    * Many ticks could be right on top of each other, we can't tell
***
## Histogram
* if we **count** how many ticks there are per various x-range, we can create a **histogram**
* Create a **Histogram** diagram
    * We place the rug plot ticks
    * Choose a number of **bins**, we'll pick **3** (0 to 10, 10 to 20, 20 to 30)
    * Count **ticks** per **bin**
    * Create a **bar** as high as **count**
    * Y-axis can also be normalized as percent - The high will represent the percentage of ticks in one range
    * Changing number of bins shows more detail instead of general trends (However, too high, some bins are empty)
    * Seaborn also allows us to add on a KDE plot curve on top of a histogram
***
## Kernel Density Estimation
* It is a method of **estimating** a probability density function of a random variable
* In simpler terms, it is a way of estimating a continuous probability curve for a finite data simple
* The KDE could **go beyond the original data set**, as far as the curve is concerned (the beyond may true or not true depend on the data)
* The KDE curve is the stack of all normal distribution of each tick point of rug plot
* Can change the kernel and bandwidth used which can make KDE show more or less of the variance contained in the data
***

#### Rugplot - to describe the distribution of salary

```python
# because the y is interpretable => only use x
# use height= - as percentage to change the height of tick point
plt.figure(figsize=(5,8), dpi=300)
sns.rugplot(x='salary', data=df, height=0.5)
```

## The displot - this plot is high level of plot that show three options (histogram, rug, kde)
* The **distplot** is deprecated => Do not use

```python
# use bins= to show the variable
# the bins= could change due to the analogous parameter in kde - bandwidth
sns.displot(data=df, x="salary", bins=100) # return back histogram
```

```python
# change the style of histogram
# this is the high level of seaborn => if wrong run all for reset
sns.set(style="ticks")
sns.displot(data=df, x="salary", bins=10, color="red", edgecolor="black")
```

```python
# many parameters of matplotlib also available in seaborn call
sns.displot(data=df, x="salary", bins=10, color="red", edgecolor="black", lw=4)
```

```python
# show kde
# use kde=True
# show rug
# use rug=True
sns.displot(data=df, x="salary", bins=10, color="red", edgecolor="black", lw=4, kde=True, rug=True)
```

## The histogram
* Similar like **displot**

```python
sns.histplot(data=df, x="salary", kde=True)
```

```python
sns.displot(data=df, x="salary")  # recommend
```

## Only show the KDE

```python
sns.kdeplot(data=df, x='salary')
```

### Create a example

```python
np.random.seed(42)
sample_age = np.random.randint(0, 100, 200)
```

```python
sample_age
```

```python
sample_age = pd.DataFrame(sample_age, columns=['age'])
```

```python
sample_age
```

```python
# an uniform distribution with several empties
sns.rugplot(data=sample_age, x='age')
```

```python
sns.displot(data=sample_age, x='age', rug=True, bins=30, kde=True)
```

```python
# only show the information from 0 to 100 - or avoid the further forecast of kde
# use the clip=
```

```python
sns.kdeplot(data=sample_age, x='age')
```

```python
sns.kdeplot(data=sample_age, x='age', clip=[0, 100])
```

```python
# bandwidth is the width of the variance of distribution
# change the bandwidth - show the behaviour
# large bandwith => normalize the overall normal distribution

sns.kdeplot(data=sample_age, x='age', clip=[0, 100], bw_adjust=20)
```

```python
# small bandwidth => pick up lots of the variance here
sns.kdeplot(data=sample_age, x='age', clip=[0,100], bw_adjust=0.01)
```

```python
# fill the size of kde
# using the share=True
sns.kdeplot(data=sample_age, x='age', clip=[0,100], bw_adjust=0.6, shade=True)
```
