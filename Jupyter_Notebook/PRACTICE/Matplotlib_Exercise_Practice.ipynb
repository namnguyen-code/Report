```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
mpl.rcParams.update({'figure.facecolor': 'white',
                     'font.size': 13,
                     'figure.edgecolor': 'white',
                     'axes.facecolor' : 'white',
                     'axes.edgecolor': 'black'})
```

```python
np.arange(0, 11, 1)
```

```python
m = np.linspace(0, 10, 11)
```

```python
exp_array = m * (3*10**8)**2
np.set_printoptions(formatter={'all' : lambda x: format(x,'6.3E')})
print(exp_array)
```

```python
exp_array
```

```python
plt.figure(figsize=(16,9), dpi=300)
plt.plot(m, exp_array, color='red', lw=8)
plt.xlim(0,10)
plt.title('E=mc^2', fontsize='24')
plt.xlabel('Mass in Grams')
plt.ylabel('Energy in Joules')
```

```python
plt.figure(figsize=(16,9), dpi=300)
plt.plot(m, exp_array, color='red', lw=8)
plt.grid(visible=True, axis='y')
plt.yscale('log')
plt.xlim(0,10)
plt.title('E=mc^2', fontsize='24')
plt.xlabel('Mass in Grams')
plt.ylabel('Energy in Joules')
```

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

```python
plt.figure(figsize=(16,9), dpi=300)
plt.plot(labels, july16_2007, label ='july16_2007')
plt.plot(labels, july16_2020, label = 'july16_2020')
plt.legend(bbox_to_anchor = (0.5, 0., 0.5, 0.5 )) #x, y, width, height
```

```python
plt.figure(figsize=(16,9), dpi=300)
plt.plot(labels, july16_2007, label ='july16_2007')
plt.plot(labels, july16_2020, label = 'july16_2020')
plt.legend(bbox_to_anchor = (0.5, 0., 0.7, 0.5 )) #x, y, width, height
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(16,9), dpi=300)
ax[0].plot(labels, july16_2007)
ax[0].set_title(label= 'July 16th, 2007')

ax[1].plot(labels, july16_2020)
ax[1].set_title(label = 'July 16th, 2020')
```

```python
fig1, ax1 = plt.subplots(nrows=1, ncols=1, figsize=(16,9), dpi=300)
fig1.suptitle('July 16th Yield Curves', fontsize='24')
ax1.plot(labels, july16_2007, color='blue', lw=2)
for y in ax1.get_yticklabels():
    y.set_color('blue')
ax1.spines['left'].set_color('blue')
ax1.spines['left'].set_linewidth('4')
ax1.set_ylabel(ylabel='2007', fontsize='20', color='blue')
plt.yticks(fontsize = 15)

ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, color='red', lw=2)
for labels in ax2.get_yticklabels():
    labels.set_color('red')
ax2.set_ylabel(ylabel='2020',fontsize='20', color='red')
ax2.spines['right'].set_color('red')
ax2.spines['right'].set_linewidth('4')

fig1.subplots_adjust(top = 0.88)
fig1.tight_layout()
```
