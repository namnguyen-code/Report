```python
import seaborn as sns
import matplotlib as mpl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mpl.rcParams.update({'figure.facecolor':'white',
                     'font.size': 14,
                     'axes.titlesize': 32,
                     'xtick.labelsize': 11,
                     'ytick.labelsize': 11
})
```

```python
df = pd.read_csv(filepath_or_buffer=r'C:\Users\admin\Desktop\Data Science\Course-2021\05-Seaborn\dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(16,9), dpi=300)
sns.rugplot(x='salary', data=df, height=0.5)
```

```python
sns.displot(data=df, x='salary', bins=50)
```

```python
sns.set(style='whitegrid')
sns.displot(data=df, x='salary', bins=10, color='red', edgecolor='black')
```

```python
sns.displot(data=df, x='salary', bins=10, color='red', edgecolor='black', lw=3)
```

```python
sns.displot(data=df, x='salary', bins=10, color='red', edgecolor='black', lw=3, kde=True, rug=True)
```

```python
sns.histplot(data=df, x='salary', kde=True)
```

```python
sns.set(style='ticks')
sns.kdeplot(data=df, x='salary')
```
