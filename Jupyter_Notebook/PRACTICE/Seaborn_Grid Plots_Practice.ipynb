```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.rcParams.update({'figure.facecolor': 'white',
                     'font.size': 14,
                     'xtick.labelsize': 12,
                     'ytick.labelsize': 12})
```

```python
df = pd.read_csv(r'C:\Users\admin\Desktop\Data Science\Course-2021\05-Seaborn\StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='box')
```

```python
sns.catplot(data=df, x='gender', y='math score', row='lunch', kind='box')
```

```python
sns.catplot(data=df, x='gender', y='math score', col='lunch', kind='box')
```

```python
sns.catplot(data=df, x='gender', y='math score', hue='lunch', kind='box')
```

```python
plt.figure(figsize=(16,9))
sns.catplot(data=df, x='gender', y='math score', col='test preparation course', row='lunch', kind='box', height=7)
plt.tight_layout()
```

```python
plt.figure(figsize=(16,9))
sns.catplot(data=df, x='gender', y='math score', col='lunch', row='race/ethnicity', kind='box', height=7)
plt.tight_layout()
```

```python
sns.catplot(x='gender', y='math score', data=df, row='lunch', col='test preparation course', kind='swarm', size=7)
```

```python
g = sns.PairGrid(data=df)

g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)
```

```python
a = sns.PairGrid(data=df, hue='gender')
a.map_upper(sns.scatterplot)
a.map_diag(sns.histplot)
a.map_lower(sns.kdeplot)
```
