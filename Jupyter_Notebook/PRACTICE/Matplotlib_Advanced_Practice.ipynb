```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
mpl.rcParams.update({'font.size': 13,
                     'text.color': 'black',
                     'figure.facecolor' : 'white',
                     'figure.edgecolor' : 'white',
                     'xtick.labelcolor': 'black',
                     'axes.facecolor': 'white',
                     'axes.edgecolor': 'black'
})
```

```python
x = np.linspace(0, 5, 10)
```

```python
x
```

```python
fig,ax = plt.subplots(nrows=1, ncols=2, figsize = [16, 9], dpi=300)
ax[0].plot(x, x*2, x,np.exp(x))
ax[0].set_ylim(0,160)
ax[0].set_xlim(0,5)
ax[0].set_title(label = 'Normal Scale')

ax[1].plot(x, x**2, x,np.exp(x))
ax[1].set_yscale('log') # Turn into the logarithmic scale
ax[1].set_ylim(0, 10**3)
ax[1].set_xlim(0, 5)
ax[1].set_title(label = 'Logarithmic Scale')
```

```python
fig1, ax = plt.subplots(nrows= 1, ncols= 1, dpi=300, figsize=(8,3))
ax.plot(x, x**2, x, x**3, lw=2)
xticks = [1, 2, 3, 4, 5]
ax.set_xticks(ticks=xticks)
ax.set_xticklabels(labels=[r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=13)
ax.set_xlim(0,5)

yticks = [0, 50, 100, 150]
ax.set_ylim(0,150)
ax.set_yticks(np.arange(0,151,50))
ax.set_yticks(yticks)
ax.set_yticklabels(labels=[r'$%.1f$' % y for y in yticks], fontsize=16)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,9), dpi=300)
ax.set_title(label='Science Notation')
ax.plot(x, x**2, x, np.exp(x))
from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) # Set the science notation
formatter.set_powerlimits((-1,1))
ax.yaxis.set_major_formatter(formatter)
ax.set_xlim(0, 5)
ax.set_ylim(0, 160)
```

```python
mpl.rcParams.update({'xtick.major.pad': 5,
                    'ytick.major.pad': 10})
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,4), dpi=300)
ax.plot(x, x**2, x, np.exp(x))
ax.set_title(label = 'Label and Spacing')

ax.set_xlabel("X")
ax.set_ylabel("Y")
# change the distance between the axis label and axes
ax.xaxis.labelpad = 5
ax.yaxis.labelpad = 10

ax.set_xlim(0,5)
ax.set_ylim(0,160);
```

```python
mpl.rcParams.update({'xtick.major.pad': 3,
                     'ytick.major.pad': 3})
```

```python
fig_position, ax = plt.subplots(nrows=1, ncols=1, dpi= 300)

ax.plot(x, x**2, x, np.exp(x))
ax.set_xlim(0,5)
ax.set_title(label='Label')

ax.set_xlabel("X", labelpad= 5)
ax.set_ylabel("Y", labelpad= 5)

fig_position.subplots_adjust(left=0.1, right=1.2, top=0.9, bottom =0.2)
```

```python
fig_1, ax = plt.subplots(nrows=1, ncols=2, figsize=(16,9), dpi=300)
ax[0].plot(x, x**2, x, np.exp(x))
ax[0].set_xlim(0,5)
ax[0].set_ylim(0,160)
ax[0].grid(visible=True, axis='x', linestyle = '--', markeredgecolor = 'red', linewidth = 3, markerfacecolor = 'red', color = 'red')
ax[0].grid(visible=True, axis='y', linestyle = '-.')

ax[1].plot(x, x**2, x, np.exp(x))
ax[1].set_xlim(0,5)
ax[1].set_ylim(0,160)
ax[1].grid(visible=True, axis='x', linestyle='-.', color = 'blue', linewidth = 3)

```

```python
fig_spine, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))
ax.spines['bottom'].set_color('blue')
ax.spines.left.set_color('red')
ax.spines[['top', 'right']].set_visible(False)
ax.spines[['bottom','right']].set_color('red')
ax.spines['right'].set_visible(True)
ax.spines['right'].set_position(('axes', 0.5))
ax.spines[:].set_linewidth(2)
```

```python
fig_zero, ax1 = plt.subplots(nrows=1, ncols=1, figsize= (10,5))
ax1.plot(x, x**2, linewidth = '4', color='blue')
ax1.set_ylabel(ylabel='area($m^2$)', color='blue', fontsize=24)
for y in ax1.get_yticklabels():
    y.set_color('blue')

ax2 = ax1.twinx()
ax2.plot(x,x**3, linewidth = '4', color='red')
ax2.set_ylabel(ylabel='volumn ($m^3$)', fontsize =24, color='red')
for label in ax2.get_yticklabels():
    label.set_color('red')
```
