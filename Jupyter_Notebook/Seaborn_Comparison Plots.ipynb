# Comparision Plots
## Part one: Understand the plots
* Comparison plots are essentially 2 dimensional versions of the plots we've learned about so far
* The two main plots types discussed here:
    * jointplot()
    * pairplot()
* jointplot()
    * We can map histograms to each feature of a scatter plot to clarify the distributions within each feature
    * We can also adjust the scatterplot to be a hex plot or 2D KDE plot
    * You have different options to display the relationship inside the square
        * Histogram with hexagons
            * Hexagons are dark the more points fall into their area
            * Useful when many point overlap
        * 2D KDE plots show shaded distribution between both KDEs
* pairplot() - very useful
    * The pairplot() is a quick way to compare all numerical columns in a DataFrame
    * It automatically creates a histogram for each column and a scatter plot comparison between all possible combinations of columns
    * **Warning**
        * pairplot() can be CPU and RAM intensive for large DataFrame with many columns
        * It is good idea to first filter down to only columns you are interested in

#### Jointplot

```python
# show the histogram of each variable and scatter plot
sns.jointplot(data=df, x="math score", y ="reading score")
```

```python
# show the hex rather than the scatter
# using the hex
sns.jointplot(data=df, x='math score', y= "reading score", kind="hex")
```

```python
# change the alpha to get the scatter overlap point more clear
sns.jointplot(data=df, x="math score", y="reading score", kind="scatter", alpha=0.2)
```

```python
# show the hist rather than the scatter
sns.jointplot(data=df, x="math score", y="reading score", kind="hist", palette="PuBu")
```

```python
# show the kde rather than scatter
# this type of kde also plot for each variable instead of histogram
sns.jointplot(data=df, x="math score", y="reading score", kind="kde")
```

```python
# show the shade of the kde => make kde 2D
# using shade=
sns.jointplot(data=df, x="math score", y="reading score", kind="kde", shade=True, palette="Reds")
```

```python
# Add the sub-categorical in the jointplot
# change the hue
sns.jointplot(data=df, x="math score", y="reading score", kind="scatter", hue="gender")
```

#### Pairplot


```python
# show the histogram of each variable
# show the relationship between the variables
sns.pairplot(data=df)
```

```python
# when you add the hue into the pairplot => automatically change the histogram into the kde
sns.pairplot(data=df, hue="gender")
```

```python
# still want to remain the histogram
sns.pairplot(data=df, hue="gender", diag_kind="hist")
```

```python
# Get rid of duplicate plots
# using the corner
sns.pairplot(data=df, hue="gender", corner=True)
```
