## TASK: Recreate the Categorical Plot shown below:

This plot shows information only for the bottom half of income earners in the data set. It shows the boxplots for each category of NAME_FAMILY_STATUS column for displaying their distribution of their total income. The hue is the "FLAG_OWN_REALTY" column. Note: You will need to adjust or only take part of the dataframe before recreating this plot.



```python
df = df.sort_values(by="AMT_INCOME_TOTAL", axis=0, ascending=False)
```

```python
df
```

```python
df2 = df[round(len(df)/2):]
```

```python
df2
```

```python
plt.figure(figsize=(10,8), dpi=300)
plt.title(label="Income Totals per Family Status for Bottom Half of Earners", size=16)
sns.boxplot(x="NAME_FAMILY_STATUS", y="AMT_INCOME_TOTAL", data=df2, hue="FLAG_OWN_REALTY", order=["Single / not married",
                                                                                                  "Civil marriage", "Married", "Separated",
                                                                                                  "Widow"])

plt.legend(title="FLAG_OWN_REALTY",bbox_to_anchor = (1.05, 0.5), loc=2)
```

```python
# another way to do
df3 = df.nsmallest(n=int(len(df)/2), columns="AMT_INCOME_TOTAL")
plt.figure(figsize=(10,8), dpi=300)
plt.title(label="Income Totals per Family Status for Bottom Half of Earners", size=16)
sns.boxplot(x="NAME_FAMILY_STATUS", y="AMT_INCOME_TOTAL", data=df3,
            hue="FLAG_OWN_REALTY")
plt.legend(title="FLAG_OWN_REALTY",bbox_to_anchor = (1.05, 0.5), loc=2)
```
