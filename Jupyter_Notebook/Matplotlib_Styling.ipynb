# Legends

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# add the legend that will tell me what each color represents
# first the create the label
ax.plot(x,x, label='X vs X')

ax.plot(x,x**2, label='X vs X^2')
# second the using the legend()
# loc=: the location of the legend - str or pair of floats
# ax.legend(loc='lower left')
# ax.legend(loc=0) or 'best' # default
# ax.legend(loc=4) # left corner

# place outside the figure - use the tuple similar as (0,0) in figure
# use the number more than 1 to make the legend outside the figure
ax.legend(loc=(1.1, 0.5)) # the start point 110% width, 50% height
ax.legend(loc=(0.5, 0.5))
ax.legend(loc=(-0.5, 0.5)) # the start point -50 width, 50% height
```

# Visual styling

```python
fig1 = plt.figure()

ax1 = fig1.add_axes([0,0,1,1])

# change the color of axes
# ax1.plot(x,x, color='blue')
# ax1.plot(x,x, color='purple')

# change the custom color
# using RGB Hex code
# ax1.plot(x,x, color='#650dbd', label='X vs X')
# ax1.plot(x,2**x, color='#AC371E', label='X vs X ** 2')

# adding the line with - 2 ways
# 1st way
# lw=1 - default value
# lw=10 - 10 times thickness as default
# 2nd way
# use the linewidth

# ax1.plot(x,x, color='blue', label='X vs X', lw=10)
# ax1.plot(x, x**2, color='red', label='X vs X ** 2', linewidth=0.5)

# adding the line type
# possible linetype options '--', '-', '-.', ':', 'steps'
# two ways - using 'ls' or 'linestyle'
ax1.plot(x,x, color='blue', label='X vs X', lw=10,ls='--')
ax1.plot(x, x**2, color='red', label='X vs X ** 2', linewidth=0.5, linestyle='-.')
```

```python
# custom the line style
fig2 = plt.figure()

ax2 = fig2.add_axes([0,0,1,1])
# custom the linestyle
lines = ax2.plot(x, x + 2, color='red', label='X vs X + 2', linewidth=5)
# access to the the first lines of 'line' list
# set the dashes - how many points you want and how many points you want to space after it
# 5 points, 2 spaces, 10 points, 2 spaces
lines[0].set_dashes([5,2,10,2])
```

```python
# using the marker
# axes is not the continues line due to the value of x is not continues
# want to plot each of those points that's actually occuring
# using the marker to show it off
fig2 = plt.figure()

ax2 = fig2.add_axes([0,0,1,1])
# use the documentation of matplotlib marker to analyze the available
# change the size of marker by using the 'ms' or 'markersize'
ax2.plot(x, x+1, color='purple', linewidth=2, marker='o', ls='--', ms=10)
ax2.plot(x, x*2, color='blue', linewidth=2, marker='P', ls='-.', markersize=10)
```

```python
# edit the marker edge color and marker face color
fig3 = plt.figure()

ax3 = fig3.add_axes([0,0,1,1])
# using the markercolor and markeredgewidth
# using the markeredgewidth to edge of the marker larger
# check out the matplotlib gallery library
ax3.plot(x, x+1, color='purple', linewidth=2, marker='o', ls='--', ms=10,
         markerfacecolor='red', markeredgewidth=4, markeredgecolor='orange')
```
