# Seaborn Grid
* Seaborn grid calls use Matplotlib subplots() to automatically create a grid based off a categorical column
* Instead of passing in a specific number of cols or rows for the subplots, we can simply pass in the name of the column and seaborn will automatically map the subplot
* Many of seaborn's built-in plot calls are running on top of this grid system
* Directly calling the grid system allows users to heavily customize plots

## Catplot - using it when you want to add more row and column

```python
# Use the catplot to plot as you want
# You need to specify the information of the plot
sns.catplot(data=df, x="gender", y="math score", kind="box")
```

```python
# if you add the row with your column name => it will create the boxplot for separate lunch type
# each row will based on lunch type
# This type is similar function with hue
sns.catplot(data=df, x="gender", y="math score", kind="box", row="lunch")
```

```python
# each column will based on lunch type
sns.catplot(data=df, x="gender", y="math score", kind="box", col="lunch")
```

```python
# with hue
sns.catplot(data=df, x="gender", y="math score", kind="box", hue="lunch")
```

```python
# the combination of row and column
# the row needs to different wil column

sns.catplot(data=df, x="gender", y="math score", kind="box", row="lunch", col="test preparation course", size=7)
```

```python
sns.catplot(data=df, x="gender", y="math score", kind="box", col="lunch", row="race/ethnicity", size=7)
```

### There are many type of `kind` plot that you could change in you catplot


```python
sns.catplot(data=df, x="gender", y="math score", kind="violin", row="lunch", col="test preparation course", size=7)
```

```python
sns.catplot(data=df, x="gender", y="math score", kind="swarm", row="lunch", col="test preparation course", size=7)
```

### The pair plot is the power grid functionality



```python
g = sns.PairGrid(data=df)
# three kinds of mapping (upper mapping, lower mapping, diagonal)
g = g.map_upper(sns.scatterplot) # use the scatter plot for the upper of the subplot

# remember the diagonal plot is only work with one-dimension plot
g = g.map_diag(sns.histplot) # not work with scatter

g = g.map_lower(sns.kdeplot) # use the kde plot for the lower of the subplot
```

```python
# Using hue in the Pair Grid
a = sns.PairGrid(data=df, hue="gender")

a = a.map_upper(sns.scatterplot)

a = a.map_diag(sns.histplot)

a = a.map_lower(sns.kdeplot)

# should add legend to clear the plot
a = a.add_legend()
```
