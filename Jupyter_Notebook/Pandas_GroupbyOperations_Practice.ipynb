```python
# Answer the question how much the 'mpg' change per year?
# Using the groupby() - the groupby() is the lazy function
# The groupby() don't do anything until the aggregate function call out
df.groupby('model_year').mean()['mpg']
```

```python
# Calling out multiple index
df.groupby(['model_year', 'cylinders']).mean()
```

```python
# Calling out the column names
df.groupby(['model_year', 'cylinders']).mean().columns
```

```python
# calling out the Multi Index
df.groupby(['model_year', 'cylinders']).mean().index

```

```python
# Calling out the column you want
df.groupby(['model_year', 'cylinders']).mean()['mpg']
```

```python
# Describe the table
df.groupby(['model_year', 'cylinders']).describe()
```

```python
# Transpose
df.groupby(['model_year', 'cylinders']).describe().transpose()
```

```python
# assign into year_cyl => cutdown the code
year_cyl = df.groupby(['model_year', 'cylinders']).mean()
```

```python
# grab 4 cylinders for the year 70 - based off exact row
# using the tuple in the year_cyl.index.levels
year_cyl.loc[(70,4)]
```

# Cross Section
* Grab something have Multiple Index
* Method Call <b>".xs"</b>


```python
# level is chosen column index
# key is the row you looking at
# key is the single value - not including two or more value
year_cyl.xs(key=70, level='model_year')
```

```python
# work with multiple key
year_cyl.loc[[70,80]]
```

```python
# access to multiple 4 cylinder
# using the cross-section
# the actual level is not show in the results - because it only has one attribute or value for that level
four_cyl = year_cyl.xs(key=4, level='cylinders')
```

```python
four_cyl
```

```python
# if you want to access to both 6 and 8 cylinders, you use the filter it first
df[df['cylinders'].isin([6,8])].groupby('cylinders').mean()
```

```python
df[df['cylinders'].isin([6,8])].groupby(['model_year', 'cylinders']).mean()
```

```python
# Swap index or level position
year_cyl.swaplevel()
```

```python
# sort the index
# it will make some sense to sort the outer level
# ascending = False => from high to low
year_cyl.sort_index(level='model_year', ascending=False)
```

```python
# return the agg function list
df.agg(['std', 'mean'])['mpg']
```

```python
# using the list to perform the aggregate functions
# the key is the column you want to perform the aggregate functions
# the value is the aggregate function
df.agg({'mpg':['std', 'mean'], 'cylinders':['max', 'mean']})
```
