# Matrix Plots
* Matrix plots are the visual equivalent of displaying a pivot table
* The matrix plot displays all the data passed in, visualizing all the numeric values in DataFrame
* Note!
    * Not every DataFrame is a valid choice for a matrix plot such as a heatmap
* The two main types of matrix plot are:
    * heatmap()
        * Visually displays the distribution of cell values with a color mapping
    * clustermap()
        * Same visual as heatmap, but first conducts hierarchical clustering to reorganize data into group
* heatmap
    * Note that a heatmap should ideally have all cells be in the same units, so the color mapping makes sense across the entire DataFrame
    * In this particular case, all values were "rates" of percentage growth or change were in the heatmap
* Seaborn also comes with the ability to automatically cluster similar groupings
### Import environment

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### Change the background

```python
df = pd.read_csv(filepath_or_buffer=r"C:\Users\admin\Desktop\Data Science\Course-2021\05-Seaborn\country_table.csv")
```

```python
df.head()
```

```python
# set the index for the table
# due to the index is number so is not yet suitable for the heatmap => change the index to countries
df = df.set_index("Countries")
```

```python
df
```

```python
# if only call the data for the entire dataframe
# it will create the problem related to the data type
# specific the life expectancy is not similar with the rate
sns.heatmap(df)
```

```python
# drop the life expectancy
sns.heatmap(df.drop("Life expectancy", axis=1))
```

```python
# adding the linewidth to create the space between the square
sns.heatmap(df.drop("Life expectancy", axis=1), linewidth=0.5)
```

```python
# if you want the number rather than the color
# using the annotate
plt.figure(dpi=300)
sns.heatmap(df.drop("Life expectancy", axis=1), linewidth=0.5, annot=True)
```

```python
# change the color of the heatmap
# using the cm
sns.heatmap(df.drop("Life expectancy", axis=1), linewidth=0.5, annot=True
            , cmap="viridis") # recommend color
```

```python
# if you want to make the color to change slowly
# using the center
# for example, if center = 40, you need to go above this to get yellow color
sns.heatmap(df.drop("Life expectancy", axis=1), linewidth=0.5, annot=True
            , cmap="viridis", center=40) # don't recommend to change this
```

```python
# grouping the similar region in by uisng the clustermap
sns.clustermap(df.drop("Life expectancy", axis=1), linewidth=0.5, annot=True
            , cmap="viridis")
```

```python
df = df.drop("Life expectancy", axis=1)
```

```python
# if you want only cluster the index => cluster by row
sns.clustermap(np.transpose(df), linewidth=0.5, annot=True
               , cmap="viridis", col_cluster=False)
```
