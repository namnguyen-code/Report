# Matplotlib Basis
* The most basic way to use Matplotlib is through the function plot calls:
    * <b> plt.plot(x,y) </b>
* These function calls are simple to use, but don't allow for very high degrees of control

* We recommend using these simple plt.plot() calls for quickly visualizing relationships and data
* Later on we will explore the more robust OOP Matplotlib Figure API

* Note
    * There are slight differences in displaying plots within a notebook versus running a python script.
    * If you are running .py scripts instead of .ipynb notebooks, you will need to add the <b>plt.show()</b> command discussed in this video

```python
# import matplotlib
# .pyplot to call out the plot portion
import matplotlib.pyplot as plt
```

* For older Jupyter environment - need to run the magic self-command matplotlib inline
* %matplotlib inline - see the image show up

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
x
```

```python
y = 2*x
```

```python
y
```

* The plt.plot() does two jobs:
    * Returning the array of matplotlib aligned to the object at some location in memory
        * [<matplotlib.lines.Line2D at 0x1a4d26f9040>] => store in memory 0x1a4d26f9040>
    * Rendering a visualization (see the plot)- if only want to see the rendering of the visualization => using the plt.show() or ;

```python
# matplotlib created a linear plot between the coordinates
# we have some point [0,0], [1,2], [2,4]... => draw a line
plt.plot(x, y)
```

```python
# if you don't see the plot => using the plt.show()
# plt.show() is also important in the
plt.plot(x,y)
plt.show()
```

```python
# using the ';' after the plt.plot()
plt.plot(x,y);
```

```python
# adding the title
# adding the ';' in the end of code to run without the output print
plt.plot(x,y)
plt.title('String Title');
```

```python
# adding the Xlabel and Ylabel
# remember to put all the code in the same cell - to connect with other code
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title');
```

```python
# Adjusting the limitation of X axis and Y axis
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
# the limitation will depends on the values of x and y
plt.xlim(0,20) # the x axis only go from 0 to 20
plt.ylim(0,20) # the y axis only go from 0 to 20
plt.title('Title');
```

```python
# the save documentation
help(plt.savefig)
```
